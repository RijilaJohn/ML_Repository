Categorical Variable

```python
def bar_plot(variable):
    # we get the features
    var = data[variable]
    # count number of categorical variable
    varValue = var.value_counts()
    
    plt.figure(figsize = (9,3))
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
categorical_var = ["Gender","Customer Type","Type of Travel","Class"]
for each in categorical_var:
    bar_plot(each)
```

Numerical Variable

```python
def plot_hist(variable):
    plt.figure(figsize = (9,3))
    plt.hist(data[variable], bins = 50)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distribution with hist".format(variable))
    plt.show()
```

```python
numerical_var = ["Age", "Flight Distance", "Departure/Arrival time convenient", "Ease of Online booking", "Gate location", "Food and drink", "Online boarding", "Seat comfort", "Leg room service", "Checkin service", "Cleanliness", "Departure Delay in Minutes", "Arrival Delay in Minutes"]
for n in numerical_var:
    plot_hist(n)
```

```python
def iqr_capping(df,variables):
  
  q1=df[variables].quantile(0.25)
  q3=df[variables].quantile(0.75)
  iqr=q3-q1
  upper_whisker=q3+(1.5*iqr)
  lower_whisker=q1-(1.5*iqr)
  return lower_whisker,upper_whisker
```

```python
lower_Flight_Distance,upper_Flight_Distance=iqr_capping(data,'Flight Distance')
print('lower Flight Distance = ',lower_Flight_Distance)
print('upper Flight Distance = ',upper_Flight_Distance)
data['Flight Distance']=np.where(data['Flight Distance']>upper_Flight_Distance,upper_Flight_Distance,
                    np.where(data['Flight Distance']<lower_Flight_Distance,lower_Flight_Distance,data['Flight Distance']))
```

```python
lower_Arrival_Delay_in_Minutes,upper_Arrival_Delay_in_Minutes=iqr_capping(data,'Arrival Delay in Minutes')
print('lower Arrival Delay in Minutes = ',lower_Arrival_Delay_in_Minutes)
print('upper Arrival Delay in Minutes = ',upper_Arrival_Delay_in_Minutes)
data['Arrival Delay in Minutes']=np.where(data['Arrival Delay in Minutes']>upper_Arrival_Delay_in_Minutes,upper_Arrival_Delay_in_Minutes,
                    np.where(data['Arrival Delay in Minutes']<lower_Arrival_Delay_in_Minutes,lower_Arrival_Delay_in_Minutes,data['Arrival Delay in Minutes']))
```

```python
lower_Departure_Delay_in_Minutes,upper_Departure_Delay_in_Minutes=iqr_capping(data,'Departure Delay in Minutes')
print('lower Departure Delay in Minutes = ',lower_Departure_Delay_in_Minutes)
print('upper Departure Delay in Minutes = ',upper_Departure_Delay_in_Minutes)
data['Departure Delay in Minutes']=np.where(data['Departure Delay in Minutes']>upper_Departure_Delay_in_Minutes,upper_Departure_Delay_in_Minutes,
                    np.where(data['Departure Delay in Minutes']<lower_Departure_Delay_in_Minutes,lower_Departure_Delay_in_Minutes,data['Departure Delay in Minutes']))
```

```python
lower_On_board_service,upper_On_board_service=iqr_capping(data,'On-board service')
print('lower On-board service = ',lower_On_board_service)
print('upper On-board service = ',upper_On_board_service)
data['On-board service']=np.where(data['On-board service']>upper_On_board_service,upper_On_board_service,
                    np.where(data['On-board service']<lower_On_board_service,lower_On_board_service,data['On-board service']))
```

```python
lower_Checkin_service,upper_Checkin_service=iqr_capping(data,'Checkin service')
print('lower Checkin service = ',lower_Checkin_service)
print('upper Checkin service = ',upper_Checkin_service)
data['Checkin service']=np.where(data['Checkin service']>upper_Checkin_service,upper_Checkin_service,
                    np.where(data['Checkin service']<lower_Checkin_service,lower_Checkin_service,data['Checkin service']))
```

```python
X=data.drop(['satisfaction','Age','Flight Distance','Arrival Delay in Minutes'],axis=1)
y=data['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
Scaler_X = StandardScaler()
X_train = Scaler_X.fit_transform(X_train)
X_test = Scaler_X.transform(X_test)
```

LOGISTIC REGRESSION

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
rfc__pred = rfc.predict(X_test)
print(accuracy_score(y_test,rfc__pred))
print(confusion_matrix(y_test,rfc__pred))
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k=[]
neighbors=np.arange(3,15)

for k in neighbors:
  classifier = KNeighborsClassifier(n_neighbors=k,metric='manhattan')
  classifier.fit(X_train,y_train)
  y_pred = classifier.predict(X_test)
  acc = accuracy_score(y_test,y_pred)
  metric_k.append(acc)
```
